```python
import numpy as np
```

```python
arr = np.arange(10)
print(arr, '\n')
print(np.sqrt(arr), '\n')
print(np.exp(arr), '\n')
print(np.max(arr))
print(np.min(arr))
arr5 = np.array( [ 2.34, 3.879, -2.34, 0.434, -0.988, 234.234, 123123.113 ] )
print(arr5,'\n')
print(np.sqrt(arr5),'\n') # quando tem valor negativo retorna 'nan'
print(np.abs(arr5),'\n')
print(np.sign(arr5),'\n') # sinal de cada elemento
print(np.ceil(arr5),'\n')
print(np.floor(arr5),'\n')
print(np.rint(arr5),'\n') # arredonda para o inteiro mais próximo -- presenva o tipo de dado

print('\n\nVerificar nan','\n')
print(np.isnan(arr5),'\n') #retorna se o valor é 'nan'
arr6 = np.sqrt(arr5)
print(arr6)
print(np.isnan(arr6),'\n')
print(arr6[~np.isnan(arr6)],'\n') #filtrando valores que não são 'nan'

print('\n\nFunção para not logico','\n')
mask = np.isnan(arr6)
print(mask)
print(np.logical_not(mask))
print(~mask,'\n') # equivalente ao logical_not
```

```python
#operações entre dois arrays
arr1 = (np.random.randn(5)*100).astype(np.int32)
arr2 = (np.random.randn(5)*100).astype(np.int32)
print(arr1,'\n')
print(arr2,'\n')
print(np.maximum(arr1,arr2),'\n')
print(np.minimum(arr1,arr2),'\n')

print(np.add(arr1, arr2), '\n')

arr3 = np.array( [ 2.34, 3.879, -2.34, 0.434, -0.988, 234.234, 123123.113 ] )
print(arr3,'\n')
frac, intei = np.modf(arr3)  #retorna a parte inteira e fracionaria de cada elemento do array
print(frac,'\n')
print(intei,'\n')

arr7 = np.arange(5)
print(arr7,'\n')
arr8 = np.array([2, 3, 2, 3, 2])
print(arr8,'\n')
print(np.power(arr7,arr8),'\n') #eleva os elementos do 1o array usando os elementos do 2o array como expoente

```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
xs, ys = np.meshgrid(x, y) #cria uma relação (x,y) -- todos de x com todos de y - matriz de coordenadas
print(xs)
print(ys)
```

```python
#logica condicional - np.where(cond, valorVerdade, valorFalso)
media = np.array([1.5, 2.0, 7.5, 8.0, 10.0, 5.4, 5.4])
mask = media >= 5
print(media)
print(mask)
print(np.where(mask, 'aprovado', 'reprovado'))
print(np.where(media>=5, 'aprovado', 'reprovado')) #sem a variavel mask

print('\n\noutro exemplo - usando arrays\n')
dados = np.array([15,50,26,10,18,26,14,9])
print(dados)
mask = dados > 18
print(mask)
vVerdade = np.array([50,55,25,15,10,20,5,90])
vFalso = np.array([11,7,42,99,6,88,24,64])
print(vVerdade)
print(vFalso)
print(np.where(mask, vVerdade, vFalso))

print('\n\noutro exemplo - array bidimensional\n')
arr2d = (np.random.randn(4,4) * 1000).astype(np.int32)
print(arr2d,'\n')
print(np.where(arr2d<=0, 0, arr2d),'\n')  #os valores negativos são substituidos por zero e os positivos são mantidos
mask = (arr2d<=0) | (arr2d>=10)
print(np.where((arr2d<=0) | (arr2d>=500) , 0, arr2d),'\n')  #valores negativos ou maiores do que 500 são substituidos por zero e os positivos são mantidos
```

```python
#estatistica
arr10 = np.abs((np.random.randn(5,4)*100).astype(np.int32))
print(arr10)
print(arr10.mean()) #calcula a media de todos os elementos
print(np.mean(arr10))
print(arr10.sum()) #soma
print(np.sum(arr10)) #soma
print(arr10.mean(axis=0)) #media ao longo das linhas -- media de cada coluna
print(arr10.mean(axis=1)) #media ao longo das colunas -- media de cada linha
print(arr10.sum(axis=0)) #soma ao longo das linhas -- soma de cada coluna
print(arr10.sum(axis=1)) #som ao longo das colunas -- som de cada linha

print('\n\nSoma Acumulada\n')
arr11 = np.abs((np.random.randn(10)*100).astype(np.int32))
print(arr11)
print(arr11.cumsum())
print(arr10.cumsum()) #soma cumulada
print(arr10.cumsum(axis=0)) #soma cumulada pelas linhas
print(arr10.cumsum(axis=1)) #soma cumulada pelas colunas

print('\n\nProduto Acumulado\n')
arr11 = np.abs((np.random.randn(10)*100).astype(np.int32))
print(arr11)
print(arr11.cumprod())
```

```python
#aplicação em booleanos --- true = 1 e false = 0
arr13 = (np.random.randn(10)*100).astype(np.int32)
print(arr13)
print(arr13>0)
print((arr13>=0).sum())
print((arr13<0).sum())

result = np.array([False, True, False, True, True])
print(result)
print(result.any()) #verifica se há algum True
print(result.all()) #verifica se todos são True
result = np.array([False, False, False, False, False])
print(result)
print(result.any()) #verifica se há algum True
print(result.all()) #verifica se todos são True
result = np.array([True, True, True, True, True])
print(result)
print(result.any()) #verifica se há algum True
print(result.all()) #verifica se todos são True
```

```python
#ordenação
arr14 = (np.random.randn(10)*100).astype(np.int32)
print(arr14)
arr14.sort()
print(arr14)
print('\n\nArray 2D\n')
arr15 = np.abs((np.random.randn(4,3)*100).astype(np.int32))
print(arr15,'\n')
arr15.sort()
print(arr15,'\n')
arr15 = np.abs((np.random.randn(4,3)*100).astype(np.int32))
print(arr15,'\n')
print('\nOrdenação pelas linhas --- colunas ordenadas --- axis=0\n')
arr15.sort(axis=0)
print(arr15,'\n')
print('\nOrdenação pelas colunas --- linhas ordenadas --- axis=1\n')
arr15 = np.abs((np.random.randn(4,3)*100).astype(np.int32))
print(arr15,'\n')
arr15.sort(axis=1)
print(arr15,'\n')
```

```python
#funções de conjuntos e unicidades
classes = np.array(['ABC', 'DEF', 'ABC', 'XYZ', 'IJK', 'ABC', 'XYZ', 'DEF', 'IJK','ABC', 'ABC'])
print(classes)
print(np.unique(classes))
print(set(classes)) #função python
print(sorted(set(classes)))  #função python

print('\n\nVerificar se elementos estão contidos em um vetor\n')
print(np.in1d(classes,['ABC','XYZ'])) #retorna um vetor com True e False

print('\n\nIntersecção\n')
print(np.intersect1d(classes,['XYZ', 'IJK'])) #retorna a insersecção ordenada

print('\n\nUnião\n')
print(np.union1d(classes,['XYZ', 'IJK'])) #retorna a uniao ordenada
print(np.union1d(['ABC', 'IJK'],['XYZ', 'IJK'])) #retorna a uniao ordenada

print('\n\nDiferença\n')
print(np.setdiff1d(classes,['XYZ', 'IJK'])) #retorna a diferença de A para B
print(np.setdiff1d(['ABC', 'IJK'],['XYZ', 'IJK'])) #retorna a diferença de A para B


print('\n\nDiferença Simetrica -- xor\n') #elementos que estão em apenas um dos conjuntos
print(np.setxor1d(classes,['XYZ', 'IJK'])) 
print(np.setxor1d(['ABC', 'IJK'],['XYZ', 'IJK']))


```
